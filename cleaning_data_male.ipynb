```python
# Import important library
import pandas as pd
import re
import numpy as np 

file_name = "M182181_male.csv"

# Read raw file
df = pd.read_csv(file_name)
print("Preview first 20 rows:")
print(df.head(20))
```

```python
# Drop the first 9 rows
df = df.drop(index=range(0, 9)).reset_index(drop=True)
print("\nPreview after dropping first 10 rows:")
print(df.head(10))
```

```python
# Preview the last 50 rows to see the file structure
print("\nPreview last 50 rows:")
print(df.tail(50))
```

```python
# Keep only the first 121 rows (based on inspection)
df = df.iloc[:121].reset_index(drop=True)
df
```

```python
# Use the first row as header
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)

print("\nData after setting header row:")
print(df.head(10))
```

```python
# Preview column name
print(df.columns)

# Standardize the column name
clean_column = []
for c in df.columns:
    text = str(c).strip()                 
    match = re.search(r"(19|20)\d{2}", text)  
    if match:
        clean_column.append(match.group(0))     
    else:
        clean_column.append(text)               

df.columns = clean_column
print(df.columns)
```

```python
# Split age group and occupation category
age_group, occupation = [], []
current_age = None

for raw_val in df["Data Series"].fillna("").astype(str):
    s = raw_val.strip()
    if s == "":
        age_group.append(np.nan)
        occupation.append(np.nan)
        continue

    if "Employed Male Residents Aged" in s and not raw_val.startswith("  "):
        current_age = s
        age_group.append(current_age)
        occupation.append("Total (All Occupations)")
    elif "All Occupation Groups, (Total Employed Male Residents)" in s:
        if current_age is None:
            current_age = "All Ages (All Employed Residents)"
        age_group.append(current_age)
        occupation.append("Total (All Occupations)")
    else:
        if current_age is None:
            current_age = "All Ages (All Employed Residents)"
        age_group.append(current_age)
        occupation.append(s)
```

```python
# Build cleaned dataset
clean_data = df.copy()
clean_data.insert(0, "Occupation", occupation)
clean_data.insert(0, "Age Group", age_group)
# Drop "Data Series" column
clean_data = clean_data.drop(columns=["Data Series"])
clean_data
```
