```python
# Combine 3 CSVs and standardize labels
#   - clean_data.csv
#   - clean_data_female.csv
#   - clean_data_male.csv

# Import library and load the csv file
import pandas as pd
df_all    = pd.read_csv("clean_data.csv")
df_female = pd.read_csv("clean_data_female.csv")
df_male   = pd.read_csv("clean_data_male.csv")
```

```python
# Add Sex column
df_all = df_all.copy()
df_all["Sex"] = "All"

df_female = df_female.copy()
df_female["Sex"] = "Female"

df_male = df_male.copy()
df_male["Sex"] = "Male"
```

```python
# Combine all csv file 
combined = pd.concat([df_all, df_female, df_male], ignore_index=True)
combined
```

```python
# Change Age Group column value
occupation_clean = {
    "Total (All Occupations)": "All Occupations"
}

combined["Occupation"] = combined["Occupation"].replace(occupation_clean)
```

```python
# Save the output file as csv
combined.to_csv("clean_data_combined.csv", index=False)
```
